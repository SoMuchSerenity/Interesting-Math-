```python
import matplotlib.pyplot as plt
import numpy as np
Max_iter = 100
```

```python
def Mandelbrot(c):
    z = 0
    n = 0
    
    while abs(z) <=2 and n<= Max_iter:
        
        z = z*z + c
        n += 1
        
    return n
```

```python
plane = []
for a in range(-100,100,1):
    a_row = []
    for b in range(-200,100,1):
        c = complex(b/100,a/100)
        a_row.append(Mandelbrot(c))

    plane.append(a_row)
```

```python
new_plane = [[101 - j for j in i]for i in plane]
len(new_plane)
```

```python
# Visually plot the state space
plt.figure(figsize=(10, 10))
cs=plt.imshow(new_plane,cmap='hot',aspect='equal')


plt.grid(which='both')
plt.show()
```

```python
def julia_set(a,b,z):
    n = 0
    c = complex(a,b)
    while abs(z) < 2 and n <= Max_iter:
        
        z = z*z + c
        
        n += 1
        
    return n
```

```python
c = np.random.uniform(-1,1)
d = np.random.uniform(-1,1)
julia_plane = []
for a in range(-100,100,1):
    a_row = []
    for b in range(-100,100,1):
        z = complex(b/100,a/100)
        a_row.append(julia_set(c,d,z))

    julia_plane.append(a_row)
```

```python
new_julia_plane = [[101 - j for j in i]for i in julia_plane]
```

```python
plt.figure(figsize=(10, 10))
cs=plt.imshow(new_julia_plane,cmap='hot',aspect='equal')


plt.grid(which='both')
plt.show()
print(complex(c,d))
```
